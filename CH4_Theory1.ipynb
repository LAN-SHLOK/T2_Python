## Write a program to characters and spaces in a given string 
Input = "This is a python tutorial"

```python
space=0
char=0
s="This is a python tutorial"
for i in range(0,len(s)):
    if(s[i]==" "):
        space+=1
    else:
        char+=1
print("spaces:",space)
print("character:",char)
```
