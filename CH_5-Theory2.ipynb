### WAP to extract list of positive even numbers from a given list

'''
`input - [-4,-2,-1,0,1,2,4,5,6,3]`
`output - [2,4,6]`
'''

````python
user=eval(input("enter a list of numbers"))
print(type(user))
newList=[]
````

````python
for i in user:
    if i%2==0 and i>0:
        newList.append(i)
print(newList)
````

````python
l=[i for i in user if i%2==0 and i>0]
print(l)
````

## sort / sorted - key function

````python
def getlast(s):
    return s[len(s)-1]
````

````python
l = ['ab','bb','zd','bc','td']
````

## WAP to find the sum of elements in a given list , returning 0 if the list is empty.Except number 13 is very unlucky, so it doesnt count it and the number coming immediately after 13 is'nt counted

```
input = [13,1,2,3,13]
output = 5
```

```
input = [13,1,2,3,13,100,5]
output = 10
```

````python
sum=0
newList=[]
inputlist=eval(input("enter a list of numbers : "))
for i,j in enumerate(inputlist):
    if j==13 or inputlist[i-1]==13:
        continue
    elif inputlist == []:
        print(0)
    else:
        newList.append(j)
print(newList)
for i in newList:
    sum+=i
print(sum)
````

# WAP to create freq count for words in a given sentence
```
input - the quick brown fox jumps over the dog

output - {'the':2,'quick':1,...}
```

````python
sentence="the quick brown fox jumps over the dog"
sentence.lower()
words=sentence.split(" ")
````

````python
count=0;
d={}
for i in words:
    d.setdefault(i,words.count(i))
````

````python
d
````

````python
inp='the quick brown fox jumps over the dog'
freq={i:inp.count(i) for i in inp.split()}
freq
````

````python
inp='the quick brown fox jumps over the dog'
d={}
l=inp.split()
for i in l:
    d[i]=d.get(i,0)+1
d
````
