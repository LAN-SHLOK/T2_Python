## WAP to find the sum of elements in a given list , returning 0 if the list is empty.Except number 13 is very unlucky, so it doesnt count it and the number coming immediately after 13 is'nt counted

```
input = [13,1,2,3,13]
output = 5
```

```
input = [13,1,2,3,13,100,5]
output = 10
```

````python
sum=0
newList=[]
inputlist=eval(input("enter a list of numbers : "))
for i,j in enumerate(inputlist):
    if j==13 or inputlist[i-1]==13:
        continue
    elif inputlist == []:
        print(0)
    else:
        newList.append(j)
print(newList)
for i in newList:
    sum+=i
print(sum)
````

# Lambda / Anonymous function
- A small anonymous function
- It can take any number of arguments,but can have only one expression
- It has no return value;infact it returns a function
- One liner
- Not reusable

   ```PYTHON  **SYNTAX**
    lambda a,b : a+b
    ```
    
    - lambda keyword
    - parameter - a and b
    - colon - its function
    - Expression - must be a valid single python expression
  

## Example

````python
a = lambda x,y:x+y
print(a(7,8))
````

````python
a = lambda  x : "even" if x%2==0 else "odd"
````
