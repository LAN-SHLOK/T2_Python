## maketrans() and translate() method

```python
## This method is used to replace a character from a string
```

```python
intab="aeiou"
outtab="12345"
mystring="This is a string function"
table=mystring.(intab,outtab)#This is used for mapping of ascii value
print(table)
print(mystring)
newstring=mystring.translate(table)#This method is used to replace input tab with output
print(newstring)
```

## Remove punctuation

```python
import string
```

```python
string.punctuation
```

```python
mystring="T&h%i?s i!s [a] st^rin?g f{unctio}n"
table=mystring.maketrans('i','$',string.punctuation)
print(table)
newstring=mystring.translate(table)
print(newstring)
```
