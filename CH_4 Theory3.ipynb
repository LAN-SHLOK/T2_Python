## Zipping

```python
t1=(1,2,3)
t2=(4,5,6)
t=zip(t1,t2)
print(t)
t=tuple(zip(t1,t2))
print(t)#The two tuple must be of same size
```

# WAP to return maxiumum even number from the given tuple
- input - (8,7,2,3,4,5,100,3,56)
- output - 100

```python
tuple1=eval(input("Enter numbers seperated by commas - "))
maxi=0
for i in tuple1:
    if i%2==0:
        if i>maxi:
            maxi=i
print(maxi)
```

# WAP to replace the special symbols in a given string to '#'
- input - l^j$i[e?t
- output - l#j#i#e#t

```python
import string
oldstr=input("Enter string : ")
symbols=string.punctuation
for i in symbols:
    oldstr=oldstr.replace(i,"#")
print(oldstr)
```
