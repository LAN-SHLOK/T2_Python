## WAP to implement the find method using function findchar in string

```python
s="Hi I am Gutkeshwar panwala and i sell dane mai hai kesar kear ka dam aur aj peg lagaye hum"
```

```python
def findchar(s,ch):
    for i,j in enumerate(s):
        if j==ch:
            print(i)
            break
        else:
            print(-1)
            break
```

```python
findchar("Hi I am Gutkeshwar panwala and i sell dane mai hai kesar kesar ka dam aur aj peg lagaye hum",'e')
```

## WAP to shift a number based on conditions

```python
#input - 12345
#shift - 1 -> output = 23451
#shift -2 -> output = 34512
#shift >= 6 -> output = 54321
```

```python
def shift(inp,shiftno):
    if shiftno<6:
        output=inp[shiftno:]+inp[:shiftno]
        print(output)
    else:
        output=inp[::-1]
        print(output)
        
```

```python
shift("12345",8)
```

## WAP to convert the first and last character of a word in a given string to upper case

- input-this is python
- output-This IS PythoN

```python
def converter(inp):
    newstring=inp.title()
    m=newstring.split(" ")
    result=""
    for i in m:
        result += i[:-1]+i[-1].upper()+" "
    print(result)
```

```python
s="this is python"
converter(s)
```
